<a id="is_in_bounds"></a>
## isInBounds helper function

This function returns whether or not a location is within bounds of the grid.

* **location** int[]: the grid location to check
* **world** string[][]: the world grid to check on

**returns** 

* **inBounds** boolean: true if the location is in bounds, false otherwise

```python
def isInBounds(location, world):
    rows = len(world)
    cols = len(world[0])
    if location[0] < 0 or location[0] >= rows or location[1] < 0 or location[1] >= cols:
        return False
    return True
```

## breadth_first_search

* **world** List[List[str]]: the search area
* **start** Tuple[int, int]: the starting location of the robot
* **goal** Tuple[int, int]: the goal location in the search area

**returns** None (but prints out a nice view of the world and replaces appropriate cells using '🟨' and '🟩').

```python
def breadth_first_search(world, start, goal, moves):
    frontier = []
    frontier.append([start])
    while len(frontier) > 0:
        current = frontier.pop(0)
        if current[-1] == goal:
            for path in current[1:len(current) - 1]:
                world[path[0]][path[1]] = '🟩'
            return
        for move in moves:
            neighbor = (current[-1][0] + move[0], current[-1][1] + move[1])
            if isInBounds(neighbor, world) and (world[neighbor[0]][neighbor[1]] == '⬜' or neighbor == goal):
                frontier.append(current + [neighbor])
                if not neighbor == goal:
                    world[neighbor[0]][neighbor[1]] = '🟨'
    return
```

<a id="heuristic_function"></a>
## heuristic_function

This function returns the square of the straight line distance between two locations on a grid. It is used as the heuristic function for a_star_search.

* **current** int[]: the current grid location
* **goal** int[]: the goal grid location


**returns** 

* **distance** int: the square of the straight line distance between the two locations

```python
def heuristic_function(current, goal):
    return (goal[0] - current[0])**2 + (goal[1] - current[1])**2
```

<a id="a_star_search"></a>
## a_star_search

Description of your function. Keep function to 20 lines or less with no comments and expressive variable names.

* **world** List[List[str]]: the search area
* **start** Tuple[int, int]: the starting location of the robot
* **goal** Tuple[int, int]: the goal location in the search area
* **moves** List[Tuple[int, int]]: the allowable moves in the search area
* **heuristic_func** func: the specific heuristic function used to evaluate a location


**returns** None (but prints out a view of the world; see breadth_first_search function for details).

```python
def a_star_search(world, start, goal, moves):
    frontier = PriorityQueue()
    frontier.put((heuristic_function(start, goal), [start]))
    while frontier:
        current = frontier.get()
        currentPath = current[1]
        if currentPath[-1] == goal:
            for path in currentPath[1:len(currentPath) - 1]:
                world[path[0]][path[1]] = '🟩'
            return
        for move in moves:
            neighbor = (currentPath[-1][0] + move[0], currentPath[-1][1] + move[1])
            if isInBounds(neighbor, world) and (world[neighbor[0]][neighbor[1]] == '⬜' or neighbor == goal):
                neighborScore = (current[0] - heuristic_function(currentPath[-1], goal) + heuristic_function(neighbor, goal) + 1)
                frontier.put((neighborScore, currentPath + [neighbor]))
                if not neighbor == goal:
                    world[neighbor[0]][neighbor[1]] = '🟨'
    return
```
